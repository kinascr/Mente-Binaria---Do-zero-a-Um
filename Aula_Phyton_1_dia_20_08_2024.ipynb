```python

# Solicitar entradas do usurio
km_percorridos = float(input("Digite a quantidade de km percorridos: "))
dias_alugados = int(input("Digite a quantidade de dias pelos quais o carro foi alugado: "))

# Definir as taxas
custo_por_dia = 60.00
custo_por_km = 0.15

# Calcular os custos
custo_total_dias = dias_alugados * custo_por_dia
custo_total_km = km_percorridos * custo_por_km
custo_total = custo_total_dias + custo_total_km

# Exibir o resultado
print(f"O custo total a pagar  R$ {custo_total:.2f}")
```

```python

cigarros_por_dia = int(input("Quantos cigarros voc fuma por dia? "))


anos_fumando = int(input("H quantos anos voc fuma? "))

# Calcula o nmero total de cigarros fumados
total_cigarros = cigarros_por_dia * (anos_fumando * 365)

# Calcula a perda de tempo de vida em minutos
perda_em_minutos = total_cigarros * 10

# Converte a perda de tempo de vida de minutos para dias
perda_em_dias = perda_em_minutos / (60 * 24)

# Exibe o resultado
print(f"Voc perder aproximadamente {perda_em_dias:.2f} dias de vida devido ao fumo.")
```
