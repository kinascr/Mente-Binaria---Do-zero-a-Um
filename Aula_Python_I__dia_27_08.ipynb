```python
numero1 = int(input('Digite o primeiro número: '))             #versão 2
numero2 = int(input('Digite o segundo número: '))
operacao = input('Digite a operação (+, -, *, /): ')


if operacao == '+':
    resultado = numero1 + numero2
    print(f'O seu resultado é: {resultado}')
elif operacao == '-':
    resultado = numero1 - numero2
    print(f'O seu resultado é: {resultado}')
elif operacao == '*':
    resultado = numero1 * numero2
    print(f'O seu resultado é: {resultado}')
elif operacao == '/':
    if numero2 != 0:
        resultado = numero1 / numero2
        print(f'O seu resultado é: {resultado}')
    else:
        print('Erro: Divisão por zero não permitida!')
else:
    print('Operação inválida! Por favor, utilize uma das operações: +, -, *, /')
```

```python

casa_comprar = float(input('Qual o valor da casa que deseja comprar? R$: '))
salario = float(input('Informe o seu salário R$: '))
anos_pagar = int(input('Em quantos anos deseja pagar: '))

# Calcula o valor da prestação mensal
meses = anos_pagar * 12
prestacao = casa_comprar / meses

# Calcula o limite de 30% do salário
limite = salario * 0.30

# Verifica se a prestação é inferior ou igual ao limite
if prestacao <= limite:
    print(f'Empréstimo aprovado! A prestação será de R$ {prestacao:.2f} por mês.')
else:
    print(f'Empréstimo negado! A prestação de R$ {prestacao:.2f} excede 30% do seu salário de R$ {limite:.2f}.')
```
