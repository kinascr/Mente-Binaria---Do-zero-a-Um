```python
numero1 = int(input('Digite o primeiro nmero: '))             #verso 2
numero2 = int(input('Digite o segundo nmero: '))
operacao = input('Digite a operao (+, -, *, /): ')


if operacao == '+':
    resultado = numero1 + numero2
    print(f'O seu resultado : {resultado}')
elif operacao == '-':
    resultado = numero1 - numero2
    print(f'O seu resultado : {resultado}')
elif operacao == '*':
    resultado = numero1 * numero2
    print(f'O seu resultado : {resultado}')
elif operacao == '/':
    if numero2 != 0:
        resultado = numero1 / numero2
        print(f'O seu resultado : {resultado}')
    else:
        print('Erro: Diviso por zero no permitida!')
else:
    print('Operao invlida! Por favor, utilize uma das operaes: +, -, *, /')
```

```python

casa_comprar = float(input('Qual o valor da casa que deseja comprar? R$: '))
salario = float(input('Informe o seu salrio R$: '))
anos_pagar = int(input('Em quantos anos deseja pagar: '))

# Calcula o valor da prestao mensal
meses = anos_pagar * 12
prestacao = casa_comprar / meses

# Calcula o limite de 30% do salrio
limite = salario * 0.30

# Verifica se a prestao  inferior ou igual ao limite
if prestacao <= limite:
    print(f'Emprstimo aprovado! A prestao ser de R$ {prestacao:.2f} por ms.')
else:
    print(f'Emprstimo negado! A prestao de R$ {prestacao:.2f} excede 30% do seu salrio de R$ {limite:.2f}.')
```
