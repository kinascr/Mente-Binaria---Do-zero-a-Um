```python
# Entrada de dados fornecida pelo usuário
numero1 = int(input('Digete o primeiro número: '))
numero2 = int(input('Digete o segundo número: '))
numero3 = int(input('Digete o terceiro número: '))

# Armazenar o maior e o menor número
maior = numero1
menor = numero1

# Determina o maior número
if numero2 > maior:
    maior = numero2
if numero3 > maior:
    maior = numero3

# Determina o menor número
if numero2 < menor:
    menor = numero2
if numero3 < menor:
    menor = numero3

# Exibe o maior e o menor número
print(f'O maior número é {maior} e o menor número é {menor}')
```

```python
# Solicita o salário do funcionário
salario = int(input('Qual o salário do funcionário? '))

# Calcula os aumentos
aumento1 = salario * 0.10
aumento2 = salario * 0.15

# Verifica o salário e aplica o reajuste correspondente
if salario > 2250:
    novo_salario = salario + aumento1 #adicionado para calcular o novo salário
    print(f'Você recebeu um reajuste de 10% do seu salário, agora receberá: R$ {novo_salario:.2f}')
elif salario <= 2250:
    novo_salario = salario + aumento2
    print(f'Você recebeu um reajuste de 15% do seu salário, agora receberá: R$ {novo_salario:.2f}')
```
